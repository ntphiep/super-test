```python
import requests
from bs4 import BeautifulSoup

# URL của trang web bạn muốn crawl
url = 'https://www.bonoss.com/product-category/wheel-spacers?product_count=60'

# Gửi yêu cầu GET đến trang web
response = requests.get(url)

# Kiểm tra nếu yêu cầu thành công
if response.status_code == 200:
    # Phân tích HTML của trang web
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Ví dụ: Lấy tiêu đề của trang
    title = soup.title.text
    print('Title:', title)
    
#     # Ví dụ: Lấy tất cả các liên kết trên trang
#     links = soup.find_all('a')
#     for link in links:
#         print(link.get('href'))
# else:
#     print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
products = []

for product in soup.find_all('li', class_='product'):
    product_info = {}
    
    # Trích xuất tên sản phẩm
    product_title = product.find('h3', class_='product-title').text.strip()
    product_info['title'] = product_title
    
    # Trích xuất giá sản phẩm
    price = product.find('span', class_='price').text.strip()
    product_info['price'] = price
    
    # Trích xuất hình ảnh sản phẩm
    img_tag = product.find('img', class_='attachment-woocommerce_thumbnail')
    if img_tag:
        product_info['image'] = img_tag.get('src')
    
    # Trích xuất link chi tiết sản phẩm
    link = product.find('a', class_='product-images').get('href')
    product_info['link'] = link

    products.append(product_info)
```
